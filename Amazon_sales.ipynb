**EXPLORATORY DATA ANALYSIS**

```python
a=sns.countplot(x='Size',data=df)
plt.show()
```

```python
a=sns.countplot(x='Size',data=df)
for bars in a.containers:
  a.bar_label(bars)
plt.show()
```

```python
Quantity=df.groupby(['Size'],as_index=False)['Qty'].sum().sort_values(by=['Qty'],ascending=False)
Quantity
```

**FINDING WHICH SIZE HAS SOLD THE MOST**

```python
b=sns.barplot(x='Size',y='Qty',data=Quantity)
plt.show()
```

**COURIER STATUS**

```python
CS=df.groupby(['Status'],as_index=False)['Qty'].sum().sort_values(by=['Status'],ascending=False)
CS
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Courier Status',data=df,hue='Status')
```

From the above graph, most of the orders are shipped to buyers succesfully






**TO SEE WHICH CATEGORY IS ON TOP**

```python
plt.figure(figsize=(10,5))
sns.histplot(data=df,x='Category',bins=5,edgecolor='Black',binwidth=5)
plt.show()
```

From the graph, we observe that T-shirt is the highest selling category

```python
df.groupby(['B2B'],as_index=False)['Qty'].sum()
```

```python
B2B_Check=df['B2B'].value_counts()
plt.figure(figsize=(10,5))
plt.pie(B2B_Check,labels=B2B_Check,autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

From the above graph, we can conclude that 99.2% of buyers are retailers and 0.8% are B2B buyers

**WHICH STATE HAS HIGHEST NUMBER OF BUYERS**

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df,x='ship-state',width=1,edgecolor='Black')
plt.xticks(rotation=90)
plt.xlabel("States")
plt.ylabel("Count")
plt.title('Distribution of States')
plt.show()
```

Hence we got the result that Maharastra has highest distributions among all states

**TOP TEN HIGHEST DISTRIBUTING STATES**

```python
top10 = df['ship-state'].value_counts().head(10)
plt.figure(figsize=(10,5))
sns.countplot(data=df[df['ship-state'].isin(top10.index)],x='ship-state',width=1,edgecolor='Black')
plt.xticks(rotation=90)
plt.xlabel("States")
plt.ylabel("Count")
plt.title('Top 10 states')
plt.show()
```

**CONCLUSION**

The data analysis reveals that the business has a significant customer base in Maharashtra state, experiences high demand for T-shirts, and sees M-Size as the preferred choice among buyers.
